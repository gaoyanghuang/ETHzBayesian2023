```python
import pickle
import hddm
import pandas as pd
import matplotlib.pyplot as plt
from kabuki.analyze import *
import pylab
```

```python
data = hddm.load_csv('/home/jovyan/work/Bayesian2023/dataset/all_10_10.csv')
data['rt']=data['res2_2']
data['response'] = data['judge2']
data['subj_idx'] = data['individual']
```

```python
# set model
model1 = hddm.HDDM (data) 

#set default parameters
model1.find_starting_values() 

# MCMC
model1.sample(1000,burn=20)

# model parameter stats
# model1.print_stats()
```

```python
# check convergence
model1.plot_posteriors(['a', 't', 'v', 'a_std'])
```

```python
# check model fitting on traces of posteriors
model1.plot_posterior_predictive(figsize=(10, 20))  
```

```python
# check within-group effect of coherence

model_coh = hddm.HDDM(data, depends_on={'v': 'thiscoh'})
model_coh.find_starting_values()  
model_coh.sample(1000, burn=100)
v_coh1, v_coh2, v_coh3, v_coh4, v_coh5, v_coh6 = model_coh.nodes_db.node[['v(0.016)',
                                                                          'v(0.032)',
                                                                          'v(0.064)',
                                                                          'v(0.128)',
                                                                          'v(0.256)',
                                                                          'v(0.512)']]
hddm.analyze.plot_posterior_nodes([v_coh1, v_coh2, v_coh3, v_coh4, v_coh5, v_coh6])
```

```python
# check between-group effect on v
model_group = hddm.HDDM(data, depends_on={'v': 'group'})
model_group.find_starting_values()  
model_group.sample(1000, burn=100)
v_OCD, v_HC= model_group.nodes_db.node[['v(OCD)','v(HC)']]
                                                                        
hddm.analyze.plot_posterior_nodes([v_OCD, v_HC])
```

```python
(v_OCD.trace() > v_HC.trace()).mean()
```
